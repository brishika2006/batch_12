```python
import pandas as pd
import numpy as np

# Simulate data
np.random.seed(42)
n = 200
df = pd.DataFrame({
    'StudyHours': np.random.normal(5, 2, n).clip(0),  # average 5 hours
    'SleepHours': np.random.normal(7, 1.5, n).clip(0),  # average 7 hours
    'ExamScore': np.random.normal(75, 10, n).clip(0, 100),  # average score
    'Gender': np.random.choice(['Male', 'Female'], n)
})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, vars=['StudyHours', 'SleepHours', 'ExamScore'], hue='Gender', palette='Set2')
plt.show()
sns.lmplot(data=df, x='StudyHours', y='ExamScore', hue='Gender', markers=['o', 's'], palette='coolwarm')
plt.show()
sns.lmplot(data=df, x='SleepHours', y='ExamScore', hue='Gender', markers=['^', 'v'], palette='Set1')
plt.show()
```
