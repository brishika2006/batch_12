```python
import pandas as pd
import numpy as np

# Simulate blood sugar levels for 200 patients
np.random.seed(42)
normal_group = np.random.normal(90, 10, 140)      # Normal range
diabetic_group = np.random.normal(160, 15, 60)    # Elevated range

blood_sugar = np.concatenate([normal_group, diabetic_group])
df = pd.DataFrame({'blood_sugar_level': blood_sugar})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set(style="whitegrid")

# KDE Plot
sns.kdeplot(data=df, x='blood_sugar_level', fill=True, color='purple', linewidth=2)
plt.title("KDE Plot: Blood Sugar Levels of 200 Patients")
plt.xlabel("Blood Sugar Level (mg/dL)")
plt.ylabel("Density")
plt.show()
```
