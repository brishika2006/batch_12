```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
n = 200
df = pd.DataFrame({
    'study_hours': np.random.normal(5, 2, n).clip(0),
    'sleep_hours': np.random.normal(7, 1.5, n).clip(0),
    'exam_score': np.random.normal(70, 10, n),
    'gender': np.random.choice(['Male', 'Female'], size=n)
})

# Add some correlation
df['exam_score'] += df['study_hours'] * 3 - df['sleep_hours'] * 2
df['exam_score'] = df['exam_score'].clip(0, 100)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set(style="whitegrid")

# 1. Pairplot
sns.pairplot(df, hue='gender', kind='reg', diag_kind='kde', markers=['o', 's'])
plt.suptitle("Pairplot: Study Hours, Sleep Hours, Exam Scores", y=1.02)
plt.show()

# 2. lmplot: Study Hours vs Exam Score
sns.lmplot(data=df, x='study_hours', y='exam_score', hue='gender',
           markers=['o', 's'], palette='Set2', aspect=1.5, scatter_kws={'alpha':0.7})
plt.title("Regression Plot using lmplot: Study Hours vs Exam Score by Gender")
plt.show()

# 3. regplot: Sleep Hours vs Exam Score
sns.regplot(data=df, x='sleep_hours', y='exam_score', scatter_kws={'alpha':0.6})
plt.title("Scatter Plot with Regression Line: Sleep Hours vs Exam Score")
plt.show()
```
