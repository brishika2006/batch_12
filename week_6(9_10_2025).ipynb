```python
import pandas as pd
import numpy as np

# Simulate blood sugar levels (mg/dL)
np.random.seed(42)
n = 200
blood_sugar = np.concatenate([
    np.random.normal(90, 10, int(n * 0.6)),   # Normal range
    np.random.normal(115, 10, int(n * 0.3)),  # Prediabetic
    np.random.normal(140, 15, int(n * 0.1))   # Diabetic
])

df = pd.DataFrame({'BloodSugar': blood_sugar})
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))

# KDE plot
sns.kdeplot(data=df, x='BloodSugar', fill=True, color='teal', bw_adjust=1)

# Add reference lines
plt.axvline(99, color='green', linestyle='--', label='Upper Normal Limit')
plt.axvline(125, color='orange', linestyle='--', label='Upper Prediabetic Limit')

plt.title('Distribution of Blood Sugar Levels (KDE)')
plt.xlabel('Blood Sugar (mg/dL)')
plt.ylabel('Density')
plt.legend()
plt.show()
```
